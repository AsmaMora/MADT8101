# Product Recommendation
Example of creating the product recommendation for the customer cluster 2

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df1 = pd.read_parquet('https://github.com/AsmaMora/MADT8101/raw/main/4.Customer_Segmentation&Product_Recommendation/transaction_cluster2.parquet')
```

```python
df1.info()
```

```python
df1.head(5)
```

```python
df1 = df1[(df1['Month'].isin(['May','June'])) & (df1['Year'] == 2023)]
```

```python
df1.isnull().any()
```

```python
df1['Member ID'].isna().sum()
```

```python
df1a = df1.dropna(subset=['Member ID'])
```

```python
df1a.info()
```

```python
CustomerID_Item_matrix1 = df1a.pivot_table(
    index='Member ID',
    columns='Product',
    values='qty',
    aggfunc='sum')
```

```python
CustomerID_Item_matrix1.shape
```

```python
CustomerID_Item_matrix1 = CustomerID_Item_matrix1.applymap(lambda x: 1 if x > 0 else 0)
```

```python
CustomerID_Item_matrix1.loc[1:].head()
```

```python
user_to_user_similarity_matrix1 = pd.DataFrame(
    cosine_similarity(CustomerID_Item_matrix1)
    )
```

```python
user_to_user_similarity_matrix1.head()
```

```python
user_to_user_similarity_matrix1.columns = CustomerID_Item_matrix1.index
user_to_user_similarity_matrix1['Member ID'] = CustomerID_Item_matrix1.index
user_to_user_similarity_matrix1 = user_to_user_similarity_matrix1.set_index('Member ID')
```

```python
user_to_user_similarity_matrix1.head()
```

```python
user_to_user_similarity_matrix1.loc[182746].sort_values(ascending=False)
```

```python
items_purchased_by_X1 = set(CustomerID_Item_matrix1.loc[182746].iloc[
    CustomerID_Item_matrix1.loc[182746].to_numpy().nonzero()].index)
items_purchased_by_X1
```

```python
items_purchased_by_Y1 = set(CustomerID_Item_matrix1.loc[184977].iloc[
    CustomerID_Item_matrix1.loc[184977].to_numpy().nonzero()].index)
items_purchased_by_Y1
```

```python
items_to_recommend_to_Y1 = items_purchased_by_X1 - items_purchased_by_Y1
```

```python
items_to_recommend_to_Y1
```

```python
df1a.loc[
    df1a['Product'].isin(items_to_recommend_to_Y1),
    ['Product']
].drop_duplicates().set_index('Product')
```
